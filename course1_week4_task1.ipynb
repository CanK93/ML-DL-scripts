# Распределение Максвелла

#### http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.maxwell.html#scipy.stats.maxwell

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math as m
import scipy.stats as stats
```

### Построение гистограммы и функции распределения вероятности для распределения Максвелла

```python
# построение гистограммы и функции распределения вероятности распределения Максвелла
x = np.linspace(0, 4, 1000)
r = stats.maxwell.rvs(size=1000)
plt.hist(r, bins=40, normed=True)
plt.plot(x, stats.maxwell.pdf(x), 'r-', lw=2)
plt.show()
```

### Построение графиков для выборочных средних n

```python
for i in [5, 10, 50, 100]:
    # генерируем набор данных по выбранному распределению
    rvs = stats.maxwell.rvs(size=[1000, i])
    # находим средние значения распределения
    means = [np.mean(value) for value in rvs]
    plt.hist(means, bins=40, normed=True)

    # находим среднее и ско теоретического распределения
    mean = stats.maxwell.mean()
    std = m.sqrt(stats.maxwell.var() / i)
    
    x = np.linspace(0, 4, 100)
    norm_rv = stats.norm(mean, std)
    # вычисляем плотность распределения
    norm_pdf = norm_rv.pdf(x)
    plt.title("n = %r, std: %r" % (i, std))
    plt.plot(x, norm_pdf, lw=2, color='r')
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.show()
    
```

## Вывод

Точность аппроксимации распределения выборочных средних нормальным с ростом числа n увеличивается. 
Из приведенных ниже графиков видно, что распределение выборочных средних описывается нормальным распределением, что соответствует определению ЦПТ.
